```python
data=data.rename(columns={'fixed acidity':'fixed_acidity','volatile acidity':'volatile_acidity','citric acid':'citric_acid','residual sugar':'residual_sugar','free sulfur dioxide':'free_sulfur_dioxide','total sulfur dioxide':'total_sulfur_dioxide'})
```

MEAN VALUES FOR THE VARIOUS COLUMNS


```python
mn_fa=data.fixed_acidity.mean()
mn_va=data.volatile_acidity.mean()
mn_ca=data.citric_acid.mean()
mn_ra=data.residual_sugar.mean()
mn_chl=data.chlorides.mean()
mn_fsd=data.free_sulfur_dioxide.mean()
mn_tsd=data.total_sulfur_dioxide.mean()
mn_dens=data.density.mean()
mn_pH=data.pH.mean()
mn_sul=data.sulphates.mean()
mn_alc=data.alcohol.mean()
mn_qua=data.quality.mean()

print('MEAN FOR THE VARIOUS COLUMNS')
print(f'fixed_acidity : {mn_fa}')
print(f'volatile_acidity : {mn_va}')
print(f'citric_acid : {mn_ca}')
print(f'residual_sugar : {mn_ra}')
print(f'chlorides : {mn_chl}')
print(f'free_sulphur_dioxide : {mn_fsd}')
print(f'total_sulfur_dioxide : {mn_tsd}')
print(f'density : {mn_dens}')
print(f'sulphates : {mn_sul}')
print(f'alcohol : {mn_alc}')
print(f'quality : {mn_qua}')


```

MEDIAN FOR THE COLUMNS

```python
md_fa=data.fixed_acidity.median()
md_va=data.volatile_acidity.median()
md_ca=data.citric_acid.median()
md_ra=data.residual_sugar.median()
md_chl=data.chlorides.median()
md_fsd=data.free_sulfur_dioxide.median()
md_tsd=data.total_sulfur_dioxide.median()
md_dens=data.density.median()
md_pH=data.pH.median()
md_sul=data.sulphates.median()
md_alc=data.alcohol.median()
md_qua=data.quality.median()

print('MEDIAN OF THE COLUMNS')
print(f'fixed_acidity : {md_fa}')
print(f'volatile_acidity : {md_va}')
print(f'citric_acid : {md_ca}')
print(f'residual_sugar : {md_ra}')
print(f'chlorides : {md_chl}')
print(f'free_sulphur_dioxide : {md_fsd}')
print(f'total_sulfur_dioxide : {md_tsd}')
print(f'density : {md_dens}')
print(f'pH : {md_pH}')
print(f'sulphates : {md_sul}')
print(f'alcohol : {md_alc}')
print(f'quality : {md_qua}')


```

MODE FOR THE COLUMNS

```python
mo_fa=data.fixed_acidity.mode()
mo_va=data.volatile_acidity.mode()
mo_ca=data.citric_acid.mode()
mo_ra=data.residual_sugar.mode()
mo_chl=data.chlorides.mode()
mo_fsd=data.free_sulfur_dioxide.mode()
mo_tsd=data.total_sulfur_dioxide.mode()
mo_dens=data.density.mode()
mo_pH=data.pH.mode()
mo_sul=data.sulphates.mode()
mo_alc=data.alcohol.mode()
mo_qua=data.quality.mode()

print('MODE FOR THE COLUMNS')
print(f'fixed_acidity : {mo_fa}')
print(f'volatile_acidity : {mo_va}')
print(f'citric_acid : {mo_ca}')
print(f'residual_sugar : {mo_ra}')
print(f'chlorides : {mo_chl}')
print(f'free_sulphur_dioxide : {mo_fsd}')
print(f'total_sulfur_dioxide : {mo_tsd}')
print(f'density : {mo_dens}')
print(f'pH : {mo_pH}')
print(f'sulphates : {mo_sul}')
print(f'alcohol : {mo_alc}')
print(f'quality : {mo_qua}')
```

CORRELATION BETWEEN COLUMNS


```python
correlation=data.corr()
data.corr()
```

The mean values that were found were used to fill or replace all the null values in the dataframe method.This is more efficient and sustainable than the dropna method which tends to reduce the number of null values droped.

```python
data['fixed_acidity']=data['fixed_acidity'].fillna(mn_fa)
data['citric_acid']=data['citric_acid'].fillna(mn_ca)
data['chlorides']=data['chlorides'].fillna(mn_chl)
data['free_sulfur_dioxide']=data['free_sulfur_dioxide'].fillna(mn_fsd)
data['total_sulfur_dioxide']=data['total_sulfur_dioxide'].fillna(mn_tsd)
data['density']=data['density'].fillna(mn_dens)
data['pH']=data['pH'].fillna(mn_pH)
data['quality']=data['quality'].fillna(mn_qua)
```
